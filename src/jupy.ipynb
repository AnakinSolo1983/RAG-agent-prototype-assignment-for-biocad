```python
from pipe import RAGAgent
import ipywidgets as widgets
from IPython.display import display, Markdown
```

```python
agent = RAGAgent()

query_input = widgets.Text(
    description="Query:",
    placeholder="What are potential targets for Alzheimer's disease?",
    layout=widgets.Layout(width="80%")
)

run_button = widgets.Button(
    description="Run RAG",
    button_style="primary"
)

output = widgets.Output()
```

```python
def run_rag(b):
    with output:
        output.clear_output()

        query = query_input.value
        if not query.strip():
            display(Markdown("âš ï¸ Please enter a query."))
            return

        answer, docs = agent.generate(query)

        display(Markdown("## ðŸ§  Generated Answer"))
        display(Markdown(answer))

        display(Markdown("## ðŸ“š Retrieved Sources"))
        for i, d in enumerate(docs, 1):
            display(Markdown(
                f"**{i}. {d['title']}**  \nPMID: {d['pmid']}"
            ))

run_button.on_click(run_rag)

display(query_input, run_button, output)
```
