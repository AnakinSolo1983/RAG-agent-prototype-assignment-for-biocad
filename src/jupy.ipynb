```python
from pipe import RAGAgent
import ipywidgets as widgets
from IPython.display import display, Markdown
```

```python
# –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –∞–≥–µ–Ω—Ç–∞ RAG
agent = RAGAgent()

# –°–æ–∑–¥–∞–Ω–∏–µ —Ç–µ–∫—Å—Ç–æ–≤–æ–≥–æ –≤–∏–¥–∂–µ—Ç–∞ –¥–ª—è –≤–≤–æ–¥–∞ –∑–∞–ø—Ä–æ—Å–∞
query_input = widgets.Text(
    description="–ó–∞–ø—Ä–æ—Å:",
    placeholder="–ö–∞–∫–æ–≤—ã –ø–æ—Ç–µ–Ω—Ü–∏–∞–ª—å–Ω—ã–µ —Ü–µ–ª–∏ –¥–ª—è –±–æ–ª–µ–∑–Ω–∏ –ê–ª—å—Ü–≥–µ–π–º–µ—Ä–∞?",
    layout=widgets.Layout(width="80%")
)

# –°–æ–∑–¥–∞–Ω–∏–µ –∫–Ω–æ–ø–∫–∏ –¥–ª—è –∑–∞–ø—É—Å–∫–∞ –∞–≥–µ–Ω—Ç–∞ RAG
run_button = widgets.Button(
    description="–ó–∞–ø—É—Å—Ç–∏—Ç—å RAG",
    button_style="primary"
)

# –°–æ–∑–¥–∞–Ω–∏–µ –≤–∏–¥–∂–µ—Ç–∞ –≤—ã–≤–æ–¥–∞ –¥–ª—è –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
output = widgets.Output()
```

```python
# –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏, –∫–æ—Ç–æ—Ä–∞—è –±—É–¥–µ—Ç –≤—ã–ø–æ–ª–Ω—è—Ç—å—Å—è –ø—Ä–∏ –Ω–∞–∂–∞—Ç–∏–∏ –∫–Ω–æ–ø–∫–∏
def run_rag(b):
    with output:
        output.clear_output()  # –û—á–∏—Å—Ç–∫–∞ –ø—Ä–µ–¥—ã–¥—É—â–µ–≥–æ –≤—ã–≤–æ–¥–∞

        query = query_input.value  # –ü–æ–ª—É—á–µ–Ω–∏–µ –∑–∞–ø—Ä–æ—Å–∞ –∏–∑ –≤–≤–æ–¥–∞
        if not query.strip():  # –ü—Ä–æ–≤–µ—Ä–∫–∞, –ø—É—Å—Ç –ª–∏ –∑–∞–ø—Ä–æ—Å
            display(Markdown("‚ö†Ô∏è –ü–æ–∂–∞–ª—É–π—Å—Ç–∞, –≤–≤–µ–¥–∏—Ç–µ –∑–∞–ø—Ä–æ—Å."))
            return

        # –ì–µ–Ω–µ—Ä–∞—Ü–∏—è –æ—Ç–≤–µ—Ç–∞ –∏ –ø–æ–ª—É—á–µ–Ω–∏–µ –¥–æ–∫—É–º–µ–Ω—Ç–æ–≤ –Ω–∞ –æ—Å–Ω–æ–≤–µ –∑–∞–ø—Ä–æ—Å–∞
        answer, docs = agent.generate(query)

        # –û—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ —Å–≥–µ–Ω–µ—Ä–∏—Ä–æ–≤–∞–Ω–Ω–æ–≥–æ –æ—Ç–≤–µ—Ç–∞
        display(Markdown("## üß† –°–≥–µ–Ω–µ—Ä–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –æ—Ç–≤–µ—Ç"))
        display(Markdown(answer))

        # –û—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö –∏—Å—Ç–æ—á–Ω–∏–∫–æ–≤
        display(Markdown("## üìö –ü–æ–ª—É—á–µ–Ω–Ω—ã–µ –∏—Å—Ç–æ—á–Ω–∏–∫–∏"))
        for i, d in enumerate(docs, 1):
            display(Markdown(
                f"**{i}. {d['title']}**  \nPMID: {d['pmid']}"
            ))

# –°–≤—è–∑—ã–≤–∞–Ω–∏–µ —Å–æ–±—ã—Ç–∏—è –Ω–∞–∂–∞—Ç–∏—è –∫–Ω–æ–ø–∫–∏ —Å —Ñ—É–Ω–∫—Ü–∏–µ–π run_rag
run_button.on_click(run_rag)

# –û—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ –≤–∏–¥–∂–µ—Ç–∞ –≤–≤–æ–¥–∞, –∫–Ω–æ–ø–∫–∏ –∏ –æ–±–ª–∞—Å—Ç–∏ –≤—ã–≤–æ–¥–∞
display(query_input, run_button, output)
```
